#Web Scrapping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Wikipedia URL for the category "Smartphones released in 2023"
category_url = "https://en.wikipedia.org/wiki/Category:Mobile_phones_introduced_in_2023"
# Send an HTTP GET request to fetch the category page content
response = requests.get(category_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the category page
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the list of pages in the category
    pages = soup.find("div", class_="mw-category").find_all("a")

    # Initialize an empty list to store data from all pages
    all_data = []

    # Loop through the pages in the category
    for page in pages:
        # Get the title of the page (the name of the smartphone)
        page_title = page.get("title")

        # Create the URL for the specific smartphone page
        page_url = f"https://en.wikipedia.org/wiki/{page_title}"

        # Send an HTTP GET request to fetch the smartphone page content
        page_response = requests.get(page_url)

        # Check if the request for the smartphone page was successful
        if page_response.status_code == 200:
            # Parse the HTML content of the smartphone page
            page_soup = BeautifulSoup(page_response.text, "html.parser")

            # Find the table containing the specifications
            table = page_soup.find("table", class_="infobox")

            # Initialize an empty dictionary to store data for this smartphone
            specifications = {}

            # Iterate through the rows of the table
            for row in table.find_all("tr"):
                label = row.find("th", class_="infobox-label")
                data = row.find("td", class_="infobox-data")

                # Check if both label and data are found
                if label and data:
                    feature = label.get_text(strip=True)
                    detail = data.get_text(strip=True)
                    specifications[feature] = detail

            # Append the specifications for this smartphone to the list
            all_data.append(specifications)

    # Create a pandas DataFrame from the scraped data
    df = pd.DataFrame(all_data)

    # Save the DataFrame to a CSV file
    df.to_csv("2023_smartphone_data.csv", index=False)

    print("Data scraped and saved to 2023_smartphone_data.csv.")
else:
    print("Failed to fetch the category page.")
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig=px.histogram(df['Price (in Rupees)'],x='Price (in Rupees)',title="Price Distribution of Smartphones")
fig.update_xaxes(tick0=20000, dtick=20000)
fig.show()
```

```python
fig = px.pie(df, values='Refresh Rate', names=df['Refresh Rate'],title="Distribution of Screen Refresh Rate",hover_data='Refresh Rate')
fig.show()
```

```python
fig = px.pie(df, values='Rear Camera (in MP)', names=df['Rear Camera (in MP)'],title="Distribution of Rear Camera Megapixel",hover_data='Rear Camera (in MP)')
fig.show()
```

```python
fig = px.pie(df, values='Front Camera (in MP)', names=df['Front Camera (in MP)'],title="Distribution of Front Camera Megapixel",hover_data='Front Camera (in MP)')
fig.show()
```

```python
df['Display Type'] = df['Display Type'].str.strip().str.upper()
fig=px.bar(df['Display Type'],x='Display Type',title="Types of Displays")
fig.show()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
df2= df.sort_values('Release Date')
# Plotting the line plot using Plotly
fig = px.line(df2, x='Release Date', y='Price (in Rupees)', title='Price over Release Date')
fig.update_xaxes(title_text='Release Date')
fig.update_yaxes(title_text='Price')
fig.show()
```

```python
plt.figure(figsize=(12, 6))
df['Manufacturer'] = df['Manufacturer'].str.strip().str.capitalize()
ax = sns.countplot(data=df['Manufacturer'], x=df['Manufacturer'], palette='Set3')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title('Number of Smartphones per Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
g = sns.PairGrid(df, hue='Manufacturer')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True,warn_singular=False)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
plt.show()
```
